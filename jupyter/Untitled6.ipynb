```ruby
require 'daru'
require 'daru_plotting_plotly'
Daru.plotting_library = :plotly
require 'httpclient'
require 'json'
```

```ruby
client = HTTPClient.new
movie_id = '4d512667b15fb2c51684c40f0668cbaf1487193576.06'
url = "https://0enrq3tgnh.execute-api.us-west-2.amazonaws.com/test/run"
response = client.post(
  url, body: ({ movie_id: movie_id }).to_json,
  header:{ 'Content-Type': 'application/json', 'x-api-key': 'VTcq6MDUIa8UMrO2WKa5G2lrKoTE1WEsaFC1fYML' })
response.body
```

```ruby
client = HTTPClient.new
url = "https://0enrq3tgnh.execute-api.us-west-2.amazonaws.com/test/tags"
response = client.get(
  url, query: { movie_id: movie_id },
  header:{ 'Content-Type': 'application/json', 'x-api-key': 'VTcq6MDUIa8UMrO2WKa5G2lrKoTE1WEsaFC1fYML' })
res = JSON.parse response.body, symbolize_names: true
body = JSON.parse res[:body]
nil
```

```ruby
data = body.sort_by do |k, v|
  k.to_i
end.map do |_,v|
  (0..21).to_a.map { |i| v.include?(i) ? i : nil }
end.transpose.to_enum.with_index.map do |value, index|
  [index, value]
end.to_h

df = Daru::DataFrame.new(data)
df.add_vector :time, df.index.to_a
df.plot(x: :time, y: (0..21).to_a).show
```
