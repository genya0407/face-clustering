```ruby
require 'daru'
require 'daru_plotting_plotly'
Daru.plotting_library = :plotly
```

```ruby
json_string =  File.open('./hoge.json', 'rb').read()
data = JSON.parse json_string, symbolize_names: true
nil
```

```ruby
time_and_tag = data[:Items].map { |item| [item[:time][:N], item[:tag][:S]] }.transpose
nil
```

```ruby
df = Daru::DataFrame.new(time: time_and_tag[0], tag: time_and_tag[1])
nil
```

```ruby
tags = data[:Items].map { |item| item[:tag][:S] }.uniq
tags_to_times = tags.map do |tag|
  items_of_tag = data[:Items].select do |item|
    item[:tag][:S] == tag
  end
  [tag, items_of_tag.map { |item| item[:time][:N].to_i }]
end.map do |tag, times|
  [tag, (1..410).to_a.map{ |i| times.include?(i) ? tag : nil }]
end.to_h
```

```ruby
df = Daru::DataFrame.new(tags_to_times)
df.add_vector :time, df.index.to_a
df.plot(x: :time, y: tags).show
```
